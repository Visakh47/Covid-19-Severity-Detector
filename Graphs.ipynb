## Importing required libraries :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Covid-19 Dataset.csv",index_col='Date',parse_dates=True)
df.head()
```

```python
df["Confirmed"].plot(figsize = (10,10), color = "k", ls = "--")
plt.xlabel("Months")
plt.ylabel("No. of Confirmed cases (in 10^7)")
plt.show()
plt.show()
```

```python
df["Recovered"].plot(figsize = (10,10), ls = "--")
plt.xlabel("Months")
plt.ylabel("No. of Recovered cases (in 10^7)")
plt.show()
```

```python
df["Active Cases"].plot(figsize = (10,10),color = "m", ls = "--")
plt.xlabel("Months")
plt.ylabel("No. of Active cases (in 10^6)")
plt.show()
```

```python
df["Deaths"].plot(figsize = (10,10), color = "r", ls = "--")
plt.xlabel("Months")
plt.ylabel("No. of Death cases")
plt.show()
```

```python
country = pd.read_csv("country_vaccinations.csv")
country.head()
```

```python
cols = ['country', 'total_vaccinations', 'iso_code', 'vaccines']
vaccine = country[cols].groupby('country').max().sort_values('total_vaccinations', ascending=False).dropna(subset=['total_vaccinations'])
vaccine = vaccine.iloc[:10]

plt.figure(figsize=(16, 7))
plt.bar(vaccine.index, vaccine.total_vaccinations, color = 'lime')

plt.xticks(rotation = 90)
plt.ylabel('Number of people vaccinated (in 10^8)')
plt.xlabel('Countries')
plt.show()
```
